```python
# Q-3 sol:

"""
The following errors are defined in the ArithmeticError class:
FloatingPointError
OverflowError
ZeroDivisionError
"""
# Example-1:
import logging
logging.basicConfig(filename="assign_9_3_1.log",level=logging.DEBUG, format="%(asctime)s %(name)s %(levelname)s %(message)s")
try:
    logging.error("Entering Try-block")
    a = 98/0 # ZeroDivisionError
except ZeroDivisionError as e:
    logging.error("Entering Except-block")
    print(f"Error : {e}")
logging.shutdown()
# Example-2:
import logging
logging.basicConfig(filename="assign_9_3_2.log",level=logging.DEBUG, format="%(asctime)s %(name)s %(levelname)s %(message)s")
import math
try:
    logging.error("Entering Try-block")
    result = math.exp(1000)
except OverflowError as e:
    logging.error("Entering Except-block")
    print(f"Error : {e}")
else:
    logging.error("Entering Else-block")
    print(result)
logging.shutdown()
```

```python
# Q-4 sol:

"""
The LookupError class in Python is a base class for exceptions that are 
raised when a lookup or indexing operation fails. It serves as a superclass 
for more specific lookup-related exceptions like "KeyError" and "IndexError".
"KeyError":
It occurs when attempting to access a dictionary using a key that does not
exist in the dictionary.
"IndexError":
It occurs when attempting to access a sequence (like a list or string) using
an index that is either too large or too small.
"""
# KeyError:
import logging
logging.basicConfig(filename="assign_9_4_1.log",level=logging.DEBUG, format="%(asctime)s %(name)s %(levelname)s %(message)s")
try:
    logging.error("Entering try-block")
    dict_1 = {1:"Anurag",2:"Pareek"}
    logging.error("Printing dictionary's specific value")
    print(dict_1[3])
except KeyError as e:
    logging.error("Handling KeyError")
    print(f"Error - KeyError : {e}")
logging.shutdown()
# IndexError:
import logging
logging.basicConfig(filename="assign_9_4_2.log",level=logging.DEBUG, format="%(asctime)s %(name)s %(levelname)s %(message)s")
try: 
    logging.error("Entering try-block")
    l = [1,2,3]
    logging.error("Printing dictionary's specific value")
    print(l[7])
except IndexError as e:
    logging.error("Handling IndexError")
    print(f"Error - IndexError : {e}")
logging.shutdown()
```
